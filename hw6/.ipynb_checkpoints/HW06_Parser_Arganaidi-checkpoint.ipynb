```python
import requests
import time
from bs4 import BeautifulSoup
import json
```

```python
url = 'http://www.banki.ru/services/responses/bank/response/'
```

```python
start_index = 10157750

rating = dict()
finished = False
counts = 0
counts_404 = 0

while not finished and counts < 2000:
    response = requests.get(url + str(start_index - counts))
    time.sleep(2)
    if response.status_code == 404:
        print('Error 404 at iter: ', counts)
        counts_404 += 1
        if counts_404 > 10:
            finished = True
            print('10 times in a row Error 404')
        continue
    bs = BeautifulSoup(response.text, 'lxml')
    try:
        name = bs.find_all("span", attrs={'class': 'bank-page-header__logo', 'data-test': 'company-header-logo', 
                              'itemprop': 'logo'})[0]['title']
        feedback = bs.find_all("div", attrs={'class': 'article-text response-page__text markup-inside-small markup-inside-small--bullet',
                                      'itemprop': 'description'})[0].text.strip()
    except IndexError:
        counts += 1
        print("Couldn't get name for iter ", counts)
        continue
    try:
        value = [x  for x in bs.find_all("div", attrs={'class': 'color-gray-gray font-size-small margin-bottom-default margin-top-small'})[0] \
                .text.strip() if x.isdigit()][0]
    except:
        value = None
    if name in rating.keys():
        rating[name].append((feedback, value))
    else:
        rating[name] = [(feedback, value), ]
    counts += 1
    counts_404 = 0
```

```python
sum(len(rating[key]) for key in rating.keys())
```

```python
data = []
for k in rating.keys():
    for v in rating[k]:
        data.append({'name': k, 'text': v[0], 'rating': v[1]})
```

```python
with open('HW06_Data_Arganaidi.json', 'w') as f:
    json.dump(data, f)
```
